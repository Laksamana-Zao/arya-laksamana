===================Ekualisasi=====================

```python
def ekualisasi(image):
    row,col=image.shape
    maxDerajat=image.max()
    keabuan = var.arange(row)
    banyakWarna = var.zeros(row)
    n = image.shape[0]*image.shape[1]
    for x in image:
        for y in x:
            banyakWarna[y]+=1
    pr = banyakWarna/n

    #mencari nilai ekualisasinya
    luas = 1/(row*col)
    sk = var.zeros(row)

    for i in range(row):
        for j in range(i+1):
            sk[i] += banyakWarna[j] * luas
        sk[i] = round(sk[i] * maxDerajat)

    ekualisasi = var.zeros(image.shape)
    for i in range(row):
        for j in range(col):
            h = image[j,i]
            ekualisasi[j,i]= sk[h]

    #menampilkan diagram
    var2.figure(1)
    var2.title('Diagram Transformasi')
    var2.xlabel('Gray')
    var2.ylabel('Pixel')
    var2.bar(keabuan,sk)

    var2.figure(2)
    var2.title('Diagram Normalisasi')
    var2.xlabel('Gray')
    var2.ylabel('Pixel')
    var2.bar(sk,banyakWarna)

    return ekualisasi
```

```python
transform = ekualisasi(grayscale)
```
