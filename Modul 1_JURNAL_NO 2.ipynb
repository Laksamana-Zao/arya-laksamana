```python
import numpy as var
import matplotlib.pyplot as var2
import cv2
from matplotlib import pyplot as var3
```

```python
img = cv2.imread('waifu.jpg')
```

```python
kernel = var.ones((3,3),var.float32)/9
dst = cv2.filter2D(img,-1,kernel)
```

```python
var2.subplot(121), var3.imshow(img), var3.title('Original')
var3.xticks([]),var3.yticks([])
var3.subplot(122),var3.imshow(dst),var3.title('Averaging')
var3.xticks([]),var3.yticks([])
var3.show()
```
