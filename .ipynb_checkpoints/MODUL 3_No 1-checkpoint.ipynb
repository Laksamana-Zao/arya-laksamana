```python
import numpy as var
import matplotlib.pyplot as var2
import cv2
from matplotlib import pyplot as var3
import math
```

```python
img = cv2.imread('waifu.jpg')
l = len(img)
```

```python
grayscale = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
var3.imshow(grayscale, cmap = "gray")
```

```python
#print(grayscale)
```

```python
index = var.zeros(len(grayscale))
for i in range (len(grayscale)):
    index[i] = i
```

```python
#print(index)
```

```python
newmatrix = var.zeros(len(grayscale))

for row in grayscale:
    for col in row:
        newmatrix[col] += 1
```

```python
#print(newmatrix)
```

```python
result = newmatrix/255*255

#print(result)
```

```python
var2.xlabel("gray")
var2.ylabel("pixel")
var2.bar(index,newmatrix)
```

===================Ekualisasi=====================

```python
pdf = var.zeros(len(grayscale))

for i in range (len(grayscale)):
    pdf[i] = newmatrix[i]/len(grayscale)
```

```python
#print(pdf)
```

```python
Sk = var.zeros(len(grayscale))

for i in range (len(grayscale)):
    for j in range(i+1):
        Sk[i] = Sk[i] + newmatrix[j]
```

```python
#print(Sk)
```

```python
Bi = var.zeros(len(grayscale))

for i in range(len(grayscale)):
    Bi[i] = round(255*(Sk[i]/256))
```

```python
#print(Bi)
```

```python
Eq = var.zeros(len(grayscale))

for i in range (len(grayscale)):
    Hvar = int(Bi[i])
    Eq[Hvar]=Eq[Hvar]+pdf[i]
```

```python
#print(Eq)
```
