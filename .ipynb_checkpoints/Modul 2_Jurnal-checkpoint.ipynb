```python
import numpy as var
import matplotlib.pyplot as var2
import cv2
from matplotlib import pyplot as var3
```

```python
img = cv2.imread('waifu.jpg')
l = len(img)
```

```python
grayscale = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
var3.imshow(grayscale, cmap = "gray")
```

```python
translation = var.zeros([360,360])
```

```python
#translasi pada x=50 y=50
x = 50
y = 50

for i in range (len(grayscale)):
    for j in range (len(grayscale)):
        rw = (i+x)
        cl = (j+y)
        translation[rw,cl] =grayscale[i][j]
var3.imshow(translation, cmap = "gray")
```

```python
rotation = var.zeros([360,360])
```

```python
#Mirror

flipped = cv2.flip(grayscale,1)
var3.title("terhadap x"), var3.imshow(flipped, cmap="gray")
```

```python
flipped = cv2.flip(grayscale,0)
var3.title("terhadap y"), var3.imshow(flipped, cmap="gray")
```
